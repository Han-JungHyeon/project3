```python
df_Oriental = pd.read_csv('/content/drive/MyDrive/Colab dataset/Oriental_painting.csv')
df_Masterpiece = pd.read_csv('/content/drive/MyDrive/Colab dataset/Masterpiece.csv')
df_Abstract = pd.read_csv('/content/drive/MyDrive/Colab dataset/Abstract.csv')
df_Contemporary_art = pd.read_csv('/content/drive/MyDrive/Colab dataset/Contemporary_art.csv')
df_Picture = pd.read_csv('/content/drive/MyDrive/Colab dataset/Picture.csv')
df_Illust = pd.read_csv('/content/drive/MyDrive/Colab dataset/Illust.csv')
```

```python
df_img_Oriental = pd.read_csv('/content/drive/MyDrive/Colab dataset/Oriental_img.csv')
df_img_Masterpiece = pd.read_csv('/content/drive/MyDrive/Colab dataset/Masterpiece_img.csv')
df_img_Abstract = pd.read_csv('/content/drive/MyDrive/Colab dataset/Abstract_img.csv')
df_img_Contemporary_art = pd.read_csv('/content/drive/MyDrive/Colab dataset/Contemporary_img.csv')
df_img_Picture = pd.read_csv('/content/drive/MyDrive/Colab dataset/Picture_img.csv')
df_img_Illust = pd.read_csv('/content/drive/MyDrive/Colab dataset/Illust_img.csv')
```

**데이터 전처리**

```python
df_Oriental.head(3)
```

```python
Oriental = pd.concat([df_Oriental, df_img_Oriental.iloc[:, 1]], axis=1)
Oriental.head(3)
```

```python
Oriental = Oriental.iloc[:, 1:]
Oriental['Price'] = Oriental['Price'].str.replace(pat=r'[ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
Oriental.head(3)
```

```python
df_Masterpiece.head(3)
```

```python
Masterpiece = pd.concat([df_Masterpiece, df_img_Masterpiece.iloc[:, 1]], axis=1)
Masterpiece.head(3)
```

```python
Masterpiece = Masterpiece.iloc[:, 1:]
Masterpiece['Price'] = Masterpiece['Price'].str.replace(pat=r'[ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
Masterpiece.head(3)
```

```python
df_Abstract.head(4)
```

```python
Abstract = pd.concat([df_Abstract, df_img_Abstract.iloc[:, 1]], axis=1)
Abstract.head(3)
```

```python
Abstract = Abstract.dropna()
Abstract = Abstract.iloc[1:, 1:]
Abstract['Price'] = Abstract['Price'].str.replace(pat=r'[ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
Abstract.head(3)
```

```python
df_Contemporary_art.head(3)
```

```python
Contemporary_art = pd.concat([df_Contemporary_art, df_img_Contemporary_art.iloc[:, 1]], axis=1)
Contemporary_art.head(3)
```

```python
Contemporary_art = Contemporary_art.dropna()
Contemporary_art = Contemporary_art.iloc[:, 1:]
Contemporary_art['Price'] = Contemporary_art['Price'].str.replace(pat=r'[ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
Contemporary_art.head(3)
```

```python
df_Picture.head(3)
```

```python
Picture = pd.concat([df_Picture, df_img_Picture.iloc[:, 1]], axis=1)
Picture.head(3)
```

```python
Picture = Picture.dropna()
Picture = Picture.iloc[:, 1:]
Picture['Price'] = Picture['Price'].str.replace(pat=r'[ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
Picture.head(3)
```

```python
df_Illust.head(3)
```

```python
Illust = pd.concat([df_Illust, df_img_Illust.iloc[:, 1]], axis=1)
Illust.head(3)
```

```python
Illust = Illust.iloc[:, 1:]
Illust['Price'] = Illust['Price'].str.replace(pat=r'[ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
Illust.head(3)
```

**데이터 합치기**

```python
Oriental_Type = np.repeat('Oriental_painting', len(Oriental))
Oriental['Type'] = Oriental_Type
Masterpiece_Type = np.repeat('Masterpiece', len(Masterpiece))
Masterpiece['Type'] = Masterpiece_Type
Abstract_Type = np.repeat('Abstract', len(Abstract))
Abstract['Type'] = Abstract_Type
Contemporary_art_Type = np.repeat('Contemporary_art', len(Contemporary_art))
Contemporary_art['Type'] = Contemporary_art_Type
Picture_Type = np.repeat('Picture', len(Picture))
Picture['Type'] = Picture_Type
Illust_Type = np.repeat('Illust', len(Illust))
Illust['Type'] = Illust_Type
```

```python
data = pd.concat([Masterpiece, Abstract, Oriental, Contemporary_art, Picture, Illust], ignore_index=True)
data
```
